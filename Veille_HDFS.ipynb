## HDFS (Hadoop Distributed File System)  

Système de fichier distribué de Hadoop Apache. 

De par sa capacité massive et sa fiabilité, HDFS est un système de stockage très adapté au Big Data. En combinaison avec YARN, ce système augmente les possibilités de gestion de données du cluster HDFS Hadoop et permet donc de traiter le Big Data efficacement. Parmi ses principales fonctionnalités, on compte la possibilité de stocker des terabytes, voire des petabytes de données.

## Structure des données 

Fonctionne selon le principe maître/esclave : les datanodes (esclaves) stockent les données tandis que le namenode (maître) localise les blocs de données. 

Chaque fichier est décomposé en blocs de taille maximale fixe (taille par défaut : 64Mo). Ces blocs seront répartis de manière redondante sur les différents data nodes. C'est le namenode qui sait comment sont décomposés les fichiers et sur quels datanodes sont stockés ces blocs.

Cette décomposition de fichiers en blocs permet à HDFS de stocker des fichiers de taille arbitrairement grande (possibilité de stocker des terabytes, voire des petabytes de données)

## Avantages :

- Capable de gérer un très grand nombre de données (Big Data).
- Très tolérant aux erreurs : les données étant stockées dans plusieurs endroits différents + distribué sur des centaines voire des milliers de serveurs
-  Evite également la congestion de réseau en privilégiant le déplacement des opérations plutôt que le déplacement des données. 
- Portabilité : il peut fonctionner sur différents types de commodity hardwares sans aucun problème de compatibilité.

## A noter :
- Les performances de HDFS sont meilleures lorsqu’il travaille sur des fichiers de taille importante (100 Mo ou plus).

- Les performances de HDFS sont meilleures lorsqu’il travaille sur un nombre "réduit" de fichiers (des millions de fichiers plutôt que des milliards de fichiers).

- Programmé en Java, nécessite un Java Runtime Environment.
